Учество имаат земено 470 партиципанти на возраст од 18 до 35 години, повеќето се женски кои имаат BMI(body mass index) од 25 до 40. Истражувањето трае 24 месеци, и во тој период учесниците треба да ги обновуваат нивните резултати на неделно ниво. Поделени се на 2 групи, група која има пристап до технологии за следење на физичка активност, и група која нема пристап до тие. Анализа на податоците се прави на секои 6 месеци за да се утврди дали има ефект ваквиот тип на технологија.Податоците се собирани во текот на 2 години од 2010 до 2012 и објавени се во 2014год. Во сите интервали резултатите покажуваат дека групата која не користи никакви уреди за следење на активностите, има позначителни резултати односно успеале да ја намалат телесната тежина повеќе отколку учесниците од другата група. Заклучокот е дека ефектот од уредите е незначителен, а во некои случаи и негативно влијае, а со тоа и просечните резултати се послаби. Сепак резултатите немора да покажуваат дека токму уредите се тие што се виновни за слабите резултати, туку може да влијае и незаинтересираноста кај некои поединци, желбата и мотивот за слабеење, стрикно држење на диетата итн.
